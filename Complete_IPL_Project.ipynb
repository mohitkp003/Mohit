## Random Grouping of Teams

```python
import pandas as pd
import random
import time
```

```python
All_Teams = {'DC':'Delhi Capitals',
            'CSK':'Chennai Super Kings',
            'RCB':'Royal Challengers Bangalore',
            'KKR':'Kolkata Knight Riders',
            'MI':'Mumbai Indians',
            'PBKS':'Punjab Kings',
            'RR':'Rajasthan Royals',
            'SRH':'Sunrisers Hyderabad'}
```

```python
temp_list = list(All_Teams.keys())

Group_A = []
Group_B = []

for _ in range(4):
    to_show = random.choices(temp_list)[0]
    temp_list.remove(to_show)
    Group_A.append(to_show)
    
for _ in range(4):
    to_show = random.choices(temp_list)[0]
    temp_list.remove(to_show)
    Group_B.append(to_show)

print(Group_A)
print(Group_B)
```

```python
# Match Results

All_Match_Results = {'Match':[],'Team 1':[],'Team 2':[],'Winner':[],'Result':[]}
```

```python
# Points Table

P_Table_Dict = {'Team':list(All_Teams.keys()),
                   'Matches':[0, 0, 0, 0, 0, 0, 0, 0],
                   'Win':[0, 0, 0, 0, 0, 0, 0, 0],
                   'Lose':[0, 0, 0, 0, 0, 0, 0, 0],
                   'Points':[0, 0, 0, 0, 0, 0, 0, 0]}

P_Table = pd.DataFrame(P_Table_Dict).sort_values('Team')


P_Table
```

## Fixtures for Matches

```python
All_Fixtures = [[i,j] for i in Group_A for j in Group_B]

All_Matches = {a+1:b for a,b in enumerate(All_Fixtures)}
print(All_Matches)

print('A - B\n'.center(28))

for num, mat in enumerate(All_Fixtures):
    print(f"Match-{num+1}: {mat}")
```

## Match Play

```python
# Game setup
Game_Level = int(input('Please choose the Game Level: (1:Easy, 2:Medium, 3:Hard, 4:Extreme) '))

Levels = {1: [0.10, 0.12, 0.13, 0.30, 0.32, 0.03],
          2: [0.18, 0.18, 0.16, 0.15, 0.15, 0.18],
          3: [0.15, 0.15, 0.15, 0.14, 0.14, 0.27],
          4: [0.14, 0.14, 0.12, 0.10, 0.10, 0.40]}

Overs = 10
Wickets = 5

opts = [0,1,2,4,6,'Out']
probs = Levels[Game_Level]



# Team1 = 'RCB'
# Team2 = 'CSK'
for match_num in range(1,len(All_Matches)+1):

    # Match & Team Details
    Match_Num = match_num
    Team1 = All_Matches[Match_Num][0]
    Team2 = All_Matches[Match_Num][1]
    Teams = [Team1,Team2]
    
    in_1_team_score = 0
    in_2_team_score = 0
    
    print(f"{Team1} Vs {Team2}".center(150),' ')
    
    #Toss
    
    batting_team = random.choices(Teams)[0]
    Teams.remove(batting_team)
    bowling_team = Teams[0]
    
    
    print(f"{batting_team} won the toss, chose to Bat first!\n\n")
    
    
    # First Inning
    print("First Inning")
    print(f"Team: {batting_team}",end = "\n")
    
    wickets_in_hand = Wickets
    for over in range(Overs):
        print(f"\n\nOver {over + 1}")
        print('|',end='')
        for ball_count in range(1,7):
            ball = random.choices(opts, probs, k=1)[0]
            if ball != 'Out':
                print(f"{ball}, ",end='')       
                in_1_team_score += ball 
    
            else:
                print("W, ",end='')
                wickets_in_hand -= 1
                if wickets_in_hand == 0:
                    break
    
        print('\b\b|')
        time.sleep(1)
        if wickets_in_hand == 0:
            break
                
    print(f"\n\n{batting_team}: {in_1_team_score}/{Wickets - wickets_in_hand} - {over+1 if (ball_count % 6 == 0) and (ball_count != 0) else over}.{ball_count} Overs")
    print('_  _  _  _  _  _  _')
    time.sleep(3)
    
    
    # Second Inning
    print(f"\nSecond Inning,\nTarget: {in_1_team_score+1}")
    print(f"Team: {bowling_team}",end = "\n\n")
    
    wickets_in_hand = Wickets
    for over in range(Overs):
        print(f"\nOver {over + 1}")
        print('|',end='')
        for ball_count in range(1,7):
            ball = random.choices(opts, probs, k=1)[0]
            if ball != 'Out':
                print(f"{ball}, ",end='')       
                in_2_team_score += ball
                if in_2_team_score > in_1_team_score:
                    break
            else:
                print("W, ",end='')
                wickets_in_hand -= 1
                if wickets_in_hand == 0:
                    break
                else:
                    continue
        print('\b\b|')
        time.sleep(1)
        if (wickets_in_hand == 0) or (in_2_team_score > in_1_team_score):
            break
            
    print(f"\n\n{bowling_team}: {in_2_team_score}/{Wickets - wickets_in_hand} - {over+1 if (ball_count % 6 == 0) and (ball_count != 0) else over}.{ball_count} Overs")            
    
    
    # Results
    Match_Tied = False
    if in_1_team_score != in_2_team_score:    
        winner_team = batting_team if in_1_team_score > in_2_team_score else bowling_team
        loser_team = batting_team if in_1_team_score < in_2_team_score else bowling_team
    else:
        Match_Tied = True
        
    Run_Margin = in_1_team_score - in_2_team_score if in_1_team_score>in_2_team_score else in_2_team_score-in_1_team_score
    Wicket_Margin = wickets_in_hand
    if in_1_team_score>in_2_team_score:
        print("\n\n---------------------------------------------")    
        m_result = f"{batting_team} won the match by {Run_Margin} Runs!"
        print(m_result,"\n\n")
    
    elif in_1_team_score == in_2_team_score:
        print("\n\n---------------------------------------------")
        m_result = f"Match Tied!"
        print(m_result,"\n\n")
    else:
        print("\n\n---------------------------------------------")
        m_result = f"{bowling_team} won the match by {Wicket_Margin} Wickets!"
        print(m_result,"\n\n")

    # Updating Match Results and Points Table
    
    All_Match_Results['Match'].append(Match_Num)
    All_Match_Results['Team 1'].append(Team1)
    All_Match_Results['Team 2'].append(Team2)
    All_Match_Results['Winner'].append(winner_team)
    All_Match_Results['Result'].append(m_result)

    t1_idx = P_Table_Dict['Team'].index(Team1)
    t2_idx = P_Table_Dict['Team'].index(Team2)

    # Update Team 1 & Team 2 Score
    P_Table_Dict['Matches'][t1_idx] += 1
    P_Table_Dict['Matches'][t2_idx] += 1

    if Match_Tied == True:
        P_Table_Dict['Points'][t1_idx] += 1
        P_Table_Dict['Points'][t2_idx] += 1        
    
    else:
        if Team1 == winner_team:
            P_Table_Dict['Win'][t1_idx] += 1
            P_Table_Dict['Lose'][t2_idx] += 1
            P_Table_Dict['Points'][t1_idx] += 2
            P_Table_Dict['Points'][t2_idx] += 0
            
        elif Team2 == winner_team:
            P_Table_Dict['Win'][t2_idx] += 1
            P_Table_Dict['Lose'][t1_idx] += 1
            P_Table_Dict['Points'][t2_idx] += 2
            P_Table_Dict['Points'][t1_idx] += 0

    P_Table = pd.DataFrame(P_Table_Dict).sort_values('Points',ascending=False)

    print(P_Table,"\n\n")
    
    d = pd.DataFrame(All_Match_Results)
    print(d.to_string(index=False))

print("\nTournament is Over!\n")

print(f"\n\nThe Four Qualifier Teams are: {list(P_Table.head(4)['Team'])[0]}, {list(P_Table.head(4)['Team'])[1]}, {list(P_Table.head(4)['Team'])[2]}, {list(P_Table.head(4)['Team'])[3]}")






```
